Basic tensory manipulations. Few ways to create tensores:

```python
z=torch.zeros(5,3)
print(z)
print(z.dtype)
```

Created a 5x3 matrix filled with zeros. 32 bit floating point numbers is default PyTorch. You can always override the default. 

```python
i = torch.ones ((5,3), dtype=torch.int16)
print(i)
```

When we do change the default, tensorFlow helps by reporting the change when printed.

It's common to initialize learning weights randomly, often with a specific seed for the PRNG for reproducibility of results.

Learning weights: Numbers neural networks learn during training. What makes a model "Smart" 

input x weight = output. Large weight, extremely important, weight zero, input ignored. 

```python
torch.manual_seed(1729)
r1 = torch.rand(2, 2)
print ("A random tensor: ")
print (r1)

r2 = torch.rand(2, 2)
print("\nA different random tensor: ")
print(r2) # new value

torch.manual_seed(1729)
r3 = torch.rand(2, 2)
print("\nThe same random tensor as before: ")
print(r3) # repeats tensor value
```

PyTorch tensors perform arithmetic operations intuitively. Tensors of similar shapes may be added, multiplied, etc. Operations with scalars are distributed over the tensor. 

```python
ones = torch.ones(2,3)
print(ones)

twos = torch.ones(2,3) *2
print (twos)

threes = ones + twos
print(threes)
print(threes.shape)

r1=torch.rand(2,3)
r2=torch.rand(3,2)
r3=r1+r2
print(r3)
```

Here's a small sample of the mathematical operations available:

```python
r = (torch.rand(2,2) - 0.5) * 2
print ("A random tensor with values between -1 and 1: ")
print(r)

#Common mathematical operations
print ("\n absolute value of r:")
print (torch.abs(r))

# as trigonometric functions
print("\n inverse sine of r:")
print(torch.asin(r))

#...and linear algebra operations like determinants and singular value decomposition
print ("\n determinant of r:")
print (torch.det(r))
print ("\n singular value decomposition of r:")
print (torch.svd(r))

#...and statistical and aggregate operations
print("\naverage and standard deviation of r:")
print(torch.std_mean(r))
print("\nMaximum value of r:")
print(torch.max(r))
```
