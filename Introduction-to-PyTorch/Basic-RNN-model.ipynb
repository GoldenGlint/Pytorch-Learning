```python
import torch
x=torch.randn(1, 10);
prev_h=torch.randn(1, 20);
W_h=torch.randn(20, 20);
W_x=torch.randn(10, 20);

i2h=torch.mm(W_x, x.t());
h2h=torch.mm(W_h, prev_h.t());
next_h=i2h+h2h;
next_h=next_h.tanh();

loss=next_h.sum();
loss.backward();
```

Simple recurring network, RNN
loss difference is correct output and expected
each tensor has history of how it came to be so it can calculate gradients for learning.
Input tensors is computed step-by-step from loss to top in reverse
W_x is what does the current data mean
H_x is what have I learned before
