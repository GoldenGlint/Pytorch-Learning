Pytorch: Software for Machine Learning
Machine Learning Framework: 
- Deep Learning primitives, NN layer types, activation & loss functions, optimizers. Also includes pretrained data models.
Research Prototyping
- Models are Python code, Autograd (pass done with single function call) and "eager mode" for dynamic model architectures
Production Deployment
- TorchScript, TorchServe, quantization
Open Source
- Large community, many users, and a lot of huge companies use it.


Tensors: Center of PyTorch. Input, Output Leraning grade. Multidimensional array with extra stuff. Bundled with 300 operations.
Accessed through a Python API, computation happens in compiled C++ code.
