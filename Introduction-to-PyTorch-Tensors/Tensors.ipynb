Creating tensors
Simples way is torch.empty() call

```python
import torch
x = torch.empty(3,4)
print(type(x))
print(x)
```

- tensor itself is 2d, having 3 rows 4 columns
- obj is tourch.Tensor, whith is alias for toch.FloatTensor; by default it's populated with 32-bit floating point
- see random-looking alues when printing your tensor, roch.empty() allocates memory for tensor, but does not initialize any values, so what is here is whatever was in memory
Brief note about tensors and number of dimensions and terminology
- see a 1d tensor called vector
- 2d tensor is a matrix
- Anything larget than 2d is just a tensor

More often than not, want to initialize tensor with some value, common cases all zeros, all ones, or random values, and the torch model provides factory methods for all of these

```python
zeros=torch.zeros(2,3)
print(zeros)
ones=torch.ones(2,3)
print(ones)

torch.manual_seed(1729)
random=torch.rand(2,3)
print(random)
```

The factor methods do all you'd expect we have a tensor full of zeros, another full of ones. Manual seed, you will reinitialize pseudo RNG and ge tthe same ones. 

```python
torch.manual_seed(1729)
random1=torch.rand(2,3)
print(random1)

random2=torch.rand(2,3)
print(random2)

torch.manual_seed(1729)
random3=torch.rand(2,3)
print(random3)

random4=torch.rand(2,3)
print(random4)
```

random 1 and random 3 are the same, random 2 and random 4 are the sa,e. Manually setting RNG seed resets, so identical computation depnding on random number should in most cases provide identical results.

Tensor Shapes
Often when performing operations on two or more tensors, need to be same shape, having same number of dimensions and cells. For that we have torch.*_like() methods.

```python
x=torch.empty(2,2,3)
print(x.shape)
print(x)

empty_like_x=torch.empty_like(x)
print(empty_like_x.shape)
print(empty_like_x)

zeros_like_x=torch.zeros_like(x)
print(zeros_like_x.shape)
print(zeros_like_x)

ones_like_x=torch.ones_like(x)
print(ones_like_x.shape)
print(ones_like_x)

rand_like_x=torch.rand_like(x)
print(rand_like_x.shape)
print(rand_like_x)
```

Last way to create a tensor that will cover is to specify its data direcly from a PyTorch collection:

```python
some_constants=torch.tensor([[3.14159, 2.71828], [1.61803, 0.57721]])
print(some_constants)

some_integers=torch.tensor((2,3,5,7,11,13,17,19))
print(some_integers)

more_integers=torch.tensor(((2,4,6), [3,6,9]))
print(more_integers)
```

Using torch.tensor() is the most straightforward way to create a tensor if you already have data in python tuple or list. Nesting them creates multi-dimensional tensor

Note: torch.tensor() creates a copy of the data
