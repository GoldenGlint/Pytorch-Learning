More often than not, want to initialize tensor with some value, common cases all zeros, all ones, or random values, and the torch model provides factory methods for all of these

```python
zeros=torch.zeros(2,3)
print(zeros)
ones=torch.ones(2,3)
print(ones)

torch.manual_seed(1729)
random=torch.rand(2,3)
print(random)
```

The factor methods do all you'd expect we have a tensor full of zeros, another full of ones. Manual seed, you will reinitialize pseudo RNG and ge tthe same ones. 

```python
torch.manual_seed(1729)
random1=torch.rand(2,3)
print(random1)

random2=torch.rand(2,3)
print(random2)

torch.manual_seed(1729)
random3=torch.rand(2,3)
print(random3)

random4=torch.rand(2,3)
print(random4)
```

random 1 and random 3 are the same, random 2 and random 4 are the sa,e. Manually setting RNG seed resets, so identical computation depnding on random number should in most cases provide identical results.

Tensor Shapes
Often when performing operations on two or more tensors, need to be same shape, having same number of dimensions and cells. For that we have torch.*_like() methods.

```python
x=torch.empty(2,2,3)
print(x.shape)
print(x)

empty_like_x=torch.empty_like(x)
print(empty_like_x.shape)
print(empty_like_x)

zeros_like_x=torch.zeros_like(x)
print(zeros_like_x.shape)
print(zeros_like_x)

ones_like_x=torch.ones_like(x)
print(ones_like_x.shape)
print(ones_like_x)

rand_like_x=torch.rand_like(x)
print(rand_like_x.shape)
print(rand_like_x)
```

Math & Logic with PyTorch Tensors
Now that you know some of the ways to create a tensor, what can you do with them?

Basic arithmetic with scalars


```python
ones=torch.zeros(2,2)+1
twos=torch.ones(2,2)*2
threes=(torch.ones(2,2)*7-1)/2
fours=twos**2;
sqrt2s=twos**0.5

print(ones)
print(twos)
print(threes)
print(fours)
print(sqrt2s)
```

Applies to every single one, addition multiplication division exponentiation distributed over every element. Output of such an operation will be tensor, you can chain them together with the usual operator precedence rule.

Simlar operations between 2 tensors behave like you'd intuitively expect:


```python
powers2=twos**torch.tensor([[1,2],[3,4]])
print(powers2)
fives=ones+fours
print(fives)
dozens=threes*fours
print(dozens)
```

Broadcasting is a way to perform an operation between tensors that have similarities in shapes. The one-row, four-column tensor is multiplied by both rows of the two-row, four column tensor. 

Way to perform operations with similarities in shapes. the one-row four column tensor is multiplied by both rows of the two-row four column tensor

This is important operation in deep learning. Multiplying a tensor of learning weights by a batch of input tensors, applying the operation to each instance in the batch seperately, and returning a tensor of identical shape. 

Rules for braodcasting are:
- Each tensor has at least one dimension
- comparing the dimension sizes of the two tensors going from last to first
- each dimension must be equal, or
- one of the dimensions must be of size 1, or
- the dimension does not exist in one of the tensors
tensors of identical shape are trivially broadcastable

```python
a= torch.ones(4,3,2)
b=a*(torch.rand(    3,2))
print(b)
c=a * torch.rand(   3,1)
print(c)
d=a*torch.rand(    1,2)
print(d)    
```

Look closely at the values of each tensor above:
-multplication that created b was broadcast over every layer of a
-for c operation braodcast over every layer and row of a
-for d switch around, every row identical across layers and columns.
