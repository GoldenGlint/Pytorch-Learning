Broadcasting is a way to perform an operation between tensors that have similarities in shapes. The one-row, four-column tensor is multiplied by both rows of the two-row, four column tensor. 

Way to perform operations with similarities in shapes. the one-row four column tensor is multiplied by both rows of the two-row four column tensor

This is important operation in deep learning. Multiplying a tensor of learning weights by a batch of input tensors, applying the operation to each instance in the batch seperately, and returning a tensor of identical shape. 

Rules for braodcasting are:
- Each tensor has at least one dimension
- comparing the dimension sizes of the two tensors going from last to first
- each dimension must be equal, or
- one of the dimensions must be of size 1, or
- the dimension does not exist in one of the tensors
tensors of identical shape are trivially broadcastable

```python
a= torch.ones(4,3,2)
b=a*(torch.rand(    3,2))
print(b)
c=a * torch.rand(   3,1)
print(c)
d=a*torch.rand(    1,2)
print(d)    
```

Look closely at the values of each tensor above:
-multplication that created b was broadcast over every layer of a
-for c operation braodcast over every layer and row of a
-for d switch around, every row identical across layers and columns.

Here are examples of broadcasting that will fail:

```python
a=torch.ones(4,3,2) 
b=a*torch.rand(4,3) #dimensions must match last to first
c=a*torch.rand(2,3) # both 2nd and 3rd dimensions are different
d=a*torch.rand((0,)) # can't broadcast with an empty tensor
```

Altering Tensors in place:

Most binary operations on tensors will return a third new tensor. new tensor c will occupy region of memory distinc. 

There are times may with to alter tensor in place. Most math funcitons have a version with appended underscore (_) that will alter a tensor in place

```python
import math
a = torch.tensor([0, math.pi/4, math.pi/2, 3*math.pi/4])
print('a:')
print(a)
print(torch.sin(a))
print(a)

b = torch.tensor([0, math.pi/4, math.pi/2, 3*math.pi/4])
print('\nb:')
print(b)
print(torch.sin_(b)) # b has changed
print(b) 
```

Arithmetic operations, there are functions that behave similarly:

```python
a=torch.ones(2,2)
b=torch.rand(2,2)

print('Before:')
print(a)
print(b)
print('\nAfter adding:')
print(a.add_(b))
print(a)
print(b)
print('\nAfter multiplying:')
print(b.mul_(b))
print(b)
```

Note that these in-place arithmetic functions are methods on the torch.Tensor object, not attached to the torch module like many other functions, eg torch.sin(). As you can see from a.add_(b), the calling tensor is the one that gets changed in place. 

There is another option for placing the result of a computation in an existing allocated tensor. Many of the functions and methods we've seen before have out argument that lets you specify tensor recieve the output if the out tensor is the correct shape and dtype. This can happen without a new memory allocation:

```python
a=torch.rand(2,2)
b=torch.rand(2,2)
c=torch.zeros(2,2)
old_id=id(c)

print(c)
d=torch.matmul(a,b,out=c)
print(c)

assert c is d
assert id(c), old_id

torch.rand(2, 2, out=c)
print(c)
assert id(c), old_id
```

Copying Tensors
As with any object in python,  assigning a tensor to a varialbe makes the variable a label of the tensor, it does not copy it.

```python
a=torch.ones(2,2)
b=a

a[0][1]=561
print(b)
```

But what if you want a seperate copy of the data to work on? The clone() method is there for you:

```python
a=torch.ones(2,2)
b=a.clone()

assert b is not a
print(torch.eq(a,b))

a[0][1]=561
print(b)
```

THere is an important thing to note when using clone(). If your source tensor has autograd enabled, then so will the clone.

Many cases, this is what you want. For example,multiple computation paths in its forward() method, and both the original tensor and its clone contribute to model output. THen to enable model learning, you want autograd turned on for both tensors. Source autograd enabled, get the result you want.

If you have a computation where neither og or clone need graidents, as long as source autograd off, good to go.

Third case: Image performing computation in forward() funciton, graidents are turned on for everything by default. want to pull some values mid stream to generate some metrics. Don't want the cloned copy of source tensor to track gradient. Performance is improved with autograd's history tracking turned off. Can use .detach() method on the source tensor:

```python
a=torch.rand(2,2,requires_grad=True)
print(a)

b=a.clone()
print(b)

c=a.detach().clone()
print(c)

print(a)
```

What's happening here?

Create a with requires_grad=true turned on. Haven't covered this optional argument yet. will during unit on autograd
print a, informs us requires_grad=True
we clone a label it b, see track computation history, inherited a autograd
clone a into c but detach it first
printing see, no computation history

Moving to GPU

One of the major advantages of PyTorch is its robust acceleration on CUDA-compadible Nvidia GPU. ("CUDA" stands for Compute Unified Device Architecture)Everything we've done has been on CPU. How do we move to that?

First we check whether a GPU is available with the is_available() method.

Note: If you do not have a CUDA-compatible GPU and CUDA drivers installed, the executable cells in this section will not execute any GPU-related code.

```python
if torch.cuda.is_available():
    print("We have a GPU!")
else:
    print("No GPU available, sorry :(")
```

Once we've determined one or more GPUs available, need someplace where GPU can see it. CPU does comutation on data in your computer's RAM. GPU has dedicated memory attached to it. Want to perform a computation on a device, must move all the data needed for computaiton to memory accessible by that device. 

Multiple ways to get your data onto target device. May do it at creation time.

```python
if torch.cuda.is_available():
    gpu_rand=torch.rand(2,2,device="cuda")
    print(gpu_rand)
else:
    print("No GPU available, sorry :(")
```

By default, new tensors are created on the CPU so we have to specify if we want to create tensor on GPU with optional device argument. See when we print the new tensor, PyTorch informs what device it's on if not CPU. 

Query # of GPUs with torch.cuda.device_count(). If you have more than one GPU, specify them by index. device='cuda:0' device='cuda:1' etc

As coding practice, specifying our degive everywhere with string contants is fragile. Ideal world, perform robustly whether on CPU or GPU hardware. Do this by creating a device handle that can be passed to tensors instead of a string:

```python
if torch.cuda.is_available():
    my_device = torch.device("cuda")
else:
    my_device = torch.device("cpu")
print("Device:{}".format(my_device))

x=torch.rand(2,2,device=my_device)
print(x)
```

If you have an existing tensor living on one device, you can move it to another with the to() method. The following line of code creates a tensor on CPI, and moves it to whichever device handle you aquired in the previous cell. 

```python
y=torch.rand(2,2)
y=y.to(my_device)
```

It is important to know that in order to do computation involving two or more tensors, all tensors must be on the same device. The following code will throw runtime error regardles of GPU device availability:

x=torch.rand(2,2)
y=torch.rand(2,2, device="gpu")
z=x+y

Manipulating Tensor Shapes:
Sometimes, you'll need to change the shape of your tensor. Blow we'll look at a few common cases and how to handle the,.

Changing the number of dimensions:
One case where you need change # of dimensions is single instance of input to your model. PyTorch models generally expect batches of input.

For example, imaging having a model that works on (3,226,226). Model expecting (N,3,226,226), where N is the number of images in the batch. So how do you make a batch of one?

```python
a = torch.rand(3,226,226)
b=a.unsqueeze(0)
print(a.shape)
print(b.shape)
```

The unsqueeze() method adds a dimension of extent 1, unsqueeze(0) adds it as a new zeroth dimension, now you have a batch of one!

Unsqueezing, what do we mean by squeezing? We're taking advantage of the fact that any dimension of extent 1 does not change the number of elements in tensor.

```python
c=torch.rand(1,1,1,1,1)
print(c)
```

continuing the example above, let's say the model's output is a 20-element vector for each input. You would then expect the output to have shape (N,20). where N is the number of instancs in the input batch. That means that for our single-input batch, we'll get an output of shape (1,20).

What if you want to do some non-batched computation with that ouput - something that's just expecting a 20-element vector?

```python
a = torch.rand(1,20)
print(a.shape)
print(a)

b=a.squeeze(0)
print(b.shape)
print(b)

c=torch.rand(2,2)
print(c.shape)

d=c.squeeze(0)
print(d.shape)
```

You can see from the shapes that our 2-dimension tensor is now 1-dimensional, and if you look closely at output of cell, a shows extra set of square brackets [] from extra dimension. Only squeeze() dimensions of extent 1. Squeeze and unsqueeze can only act on dimensions of extent 1, or else if would change # of elements in tensor. 

Another place you use unsqueeze() is ease broadcasting. Recall the example above when we had the following code:
a=torch.ones(4,3,2)
c=a*torch.rand( 3,1) #3rd dim-1, 2nd dim identical to a print(c)

The net effect of that was to broadcast operation over dimensions 0 and 2, causing random 3x1 tensor to be multiplied element-wise by every 3-elemt of column a.

Random element was just 3 element vector, lose ability to broadcast, because the final dimensions would not match up according to the broadcasting rules. unsqueeze() comes to the rescue. 

```python
a=torch.ones(4,3,2)
b=torch.rand(3)
c=b.unsqueeze(1)
print(c.shape)
print(a*c)
```

squeeze() and unsqueeze() methods also have in-place versions, squeeze_() and unsqueeze()

```python
batch_me=torch.rand(3,226,226)
print(batch_me.shape)
batch_me.unsqueeze_(0)
print(batch_me.shape)
```

Sometimes, need to change shape of tensor more radically, while still preserving elements and their contents. One case where this happens is at the interface between a convolutional layer of a model and a linera layer of the model - this is common in image classification models. A convolution kernal will yield an output tensor of shape features x width x height, but the following linear layer expects 1-d input. The reshape() will do this for you, provided that the dimensions you request yield the same number of elements as the input tensor has:

```python
output3d=torch.rand(6,20,20)
print(output3d.shape)

input1d=output3d.reshape(6*20*20)
print(input1d.shape)

#can also call it as a method on the torch module:
print(torch.reshape(output3d, (6*20*20,)).shape)
```

The (6*20*20,) argument in the final line of the cell aboe is because PyTorch expects a tuple when specifying a tensor shape - but when the shape is the first argument of a method, it lets us cheat and just use a series of integers. Here, we had to add parentheses and comma to convince the method that this is really a one-element tuple.

When it can, reshape() will return a view on the tensor to be changed, that is a seperate tensor object looking at the same underlying region of memory. Means any change made to source tnesor reflect in the view on that tensor, unless you clone() it. 

NumPy Bridge
PyTorch Broadcast semantics are compatible with NumPysm but relation is deper than that.

If you have existing ML or scientific code with data stored in NumPy ndarrays, you may wish to express that same data as PyTorch tensors, whether to take advantage of PyTorch's GPU acceleration, or its efficient abstractions for building ML models. It's easy to switch between ndarrays and PyTorch tensors:


```python
import numpy as np

numpy_array=np.ones((2,3))
print(numpy_array)

pytorch_tensor=torch.from_numpy(numpy_array)
print(pytorch_tensor)
```

PyTorch creates a tensor of the same shape containign the same data as the NumPy array, going as far to keep NumPy default 64-bit data type. Conversion can go the other way. 

```python
pytorch_rand=torch.rand(2,3)
print(pytorch_rand)

numpy_rand=pytorch_rand.numpy()
print(numpy_rand)
```

It is important to know these converted objects are using the same underlying memory as their source objects. meaning changes in one are reflected in the other. 

```python
numpy_array[1,1]=23
print(pytorch_tensor)

pytorch_rand[1,1]=17
print(numpy_rand)
```
