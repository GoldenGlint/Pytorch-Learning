Below is the ready-to-download, open access datasets from TorchVision, how to transform the images for consumption by your model, and how to use DataLoader to feed batches of data to your model.

First thing is to transform incoming images into a PyTorch tensor

```python
transform=transforms.Compose(
    [transforms.ToTensor(), 
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

Here there are two transformations for input
- transforms.ToTensor() converts images to tensors via Pillow
- transforms.Normalize centralizes around a mean of 0 and s.d of 0.5 Most activation functinons have strongest gradients at around x=0. centering speeds up learning

Networks learn faster and more reliably.
Many transforms available such as cropping, centering, rotation, reflection.

Next we'll create a CIFAR10 dataset. Set of 32x32 color tiles representing 10 classes of objects. 6 of animals (bird, cat, deer, dog, frog, horse) and 4 of vehicles (airplane, automobile, truck)

```python
trainset=torchvision.datasets.CIFAR10(root='.data', train=True, download=True, transform=transform)
```

Example of creating a dataset object in PyTorch. Downloadable datasets are subclasses of torch.utils.data.Dataset. Dataset classes in PyTorch include the downloadable datasets in TorchVision, Torchtext, and TorchAudio, as well as utility dataset classes such as torchvision.datasets.ImageFolder, which will read a folder of labeled images. You can also create your own subclasses of Dataset. The later 3 are like sublibraries that are made by the PyTorch community. 

Need to tell a few things when instantiating:
- filesystem path to where we want the data to go
- Whether or not we are using this set for trianing; most datasets will be split into training and test subsets
- whether we would like to download the dataset if we haven't already
- transformations we want to apply to that data


```python
trainloader=torch.utils.data.DataLoader(trainset, 
batch_size=4, shuffle=True, num_workers=2)
```

A Dataset subclass wraps access to the data, and is specialized to the type of data it's serving. The DataLoader knows nothing about the data, but organizes the input tensors served by the dataset into batches with the parameters you specify. 

In the example, DataLoader gives us batches of 4 images from trainset, randomizing their orders, and we told it to spin up two workers to load data from disk. It's good practice to visualize the batches your DataLoader serves. 

```python
import matplotlib.pyplot as plt
import numpy as np
classes=('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
def imshow(img):
    img=img/2+0.5 #unnormalize
    npimg=img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
#get some random training images
dataiter=iter(trainloader)
images, labels=next(dataiter)
#show images
imshow(torchvision.utils.make_grid(images))
#print labels
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
    
```
